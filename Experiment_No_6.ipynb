```python
# To calculate the Gross Salary of an employee
basic_salary = int(input("Enter the Basic Salary : "))

hra = 0
da = 0

if basic_salary <= 10000:
  hra = 0.02 * basic_salary
  da = 0.08*basic_salary
elif basic_salary <=20000:
  hra = 0.025 * basic_salary
  da = 0.09 * basic_salary
else:
  hra = 0.03 * basic_salary
  da = 0.095 * basic_salary


gross_Sal = basic_salary + hra + da

print(f"The Basic Salary is : {basic_salary}")
print(f"The HRA is : {hra}")
print(f"The DA is : {da}")
print(f"The Gross Salary of the employee is : {gross_Sal}")
```

```python
# To identify if a given character is vowel or consonent
import string
char = input("Enter the character : ")
list = ['a','e','i','o','u','A','E','I','O','U']
set1 = set(list)
lower = set(string.ascii_lowercase)
upper = set(string.ascii_uppercase)
vowel1 = lower - set1
vowel2 = upper - set1
if char in list:
  print("The Character is a Vowel")
elif char in vowel1 or char in vowel2:
  print("The character is a consonent")
else:
  print("Error! Please try again")
```
